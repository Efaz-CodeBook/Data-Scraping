```python
# Checking sample data for one Page:

all_data = data_json['courses']

all_DAAD_data = []
for item, course in enumerate(all_data):
    dates = course['date']
    
    startDate = []
    endDate = []
    eduCost = []
    regDeadline = []
    
    for date in dates:
        try:
            startD = date['start']
        except:
            startD = "Not Available"
        
        try:
            endD = date['end']
        except:
            endD = "Not Available"
        
        try:
            costS = date['costs']
        except:
            costS = "Not Available"
            
        try:
            regDead = date['registrationDeadline']
        except:
            regDead = "Not Available"
        
        startDate.append(startD)
        endDate.append(endD)
        eduCost.append(costS)
        regDeadline.append(regDead)
    d = {
            'courseType': course['courseType'],
            'courseName': course['courseName'],
            'startDate': startDate,
            'endDate': endDate,
            'eduCost': eduCost,
            'regDeadline': regDeadline
    }
    all_DAAD_data.append(d)

df = pd.DataFrame(all_DAAD_data)
df.to_csv('DAAD_Aug_23.csv', index = False)
```

```python
# Creating the data set> Just run  after importing Libraries

# Taking mother variable for all data
# Not required as already has for first page

# Inserting 'for loop' for page initiation
for page in range(100, 2200, 100):
    
    # Taking the variable for url from 2nd page
    url = f'https://www2.daad.de/deutschland/studienangebote/international-programmes/api/solr/en/search.json?cert=&admReq=&langExamPC=&scholarshipLC=&langExamLC=&scholarshipSC=&langExamSC=&fos=&langDeAvailable=&langEnAvailable=&fee=&sort=4&dur=&q=&limit=100&offset={page}&display=list&isElearning=&isSep='
    # Storing the initial response by get command
    response = requests.get(url)
    
    # Converting the response into json file
    data_json = response.json()
    
    # Taking course details only in the all data variable
    all_data = data_json['courses']
 

    # Initiating 2nd for loop for collecting each data
    for item, course in enumerate(all_data):
        dates = course['date']

        startDate = []
        endDate = []
        eduCost = []
        regDeadline = []

        # Try Except for optional data
        for date in dates:

            try:
                startD = date['start']
            except:
                startD = "Not Available"

            try:
                endD = date['end']
            except:
                endD = "Not Available"

            try:
                costS = date['costs']
            except:
                costS = "Not Available"

            try:
                regDead = date['registrationDeadline']
            except:
                regDead = "Not Available"

            startDate.append(startD)
            endDate.append(endD)
            eduCost.append(costS)
            regDeadline.append(regDead)

        # Dictionary creation for required fields
        d = {
                'courseType': course['courseType'],
                'courseName': course['courseName'],
                'subject': course['subject'],
                'academyName': course['academy'],
                'city': course['city'],
                'startDate': startDate,
                'endDate': endDate,
                'eduCost': eduCost,
                'regDeadline': regDeadline,
                'courseLanguages': course['languages'],
                'levelGerman': course['languageLevelGerman'],
                'levelEnglish': course['languageLevelEnglish'],
                'courseBeginning': course['beginning'],
                'typeCourseDate': course['typeCourseDate'],
                'costPosition': course['costString'],
                'tuitionFees': course['tuitionFees'],
                'isElearning': course['isElearning'],
                'preparationForDegree': course['preparationForDegree'],
                'prepForSubjectGroups': course['preparationForSubjectGroups'],
                'applicationDeadline': course['applicationDeadline'],
                'onlinePossible': course['isCompleteOnlinePossible'],
                'badgeLabel': course['badgeLabel'],
                'financialSupport': course['financialSupport'],
                'structuredResearch': course['structuredResearch'],
                'supportInternationalStudents': course['supportInternationalStudents'],
                'typeOfElearning': course['typeOfElearning'],
                'courseLink': course['link']
        }

        # Appending each dictionary data into the mother variable
        all_DAAD_data.append(d)
        
    # Checking the response for each page
    print(f"Page {page} is done.")


# Extracting the data into csv file

# Using pd to convert into data frame from dictionary and store in variable
df = pd.DataFrame(all_DAAD_data)

# Converting the data frame into csv file and extract
df.to_csv('DAAD_Aug_23.csv', index = False)
```
